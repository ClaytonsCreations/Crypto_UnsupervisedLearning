```python
import pandas as pd
from pathlib import Path
```

```python
#loading data
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path)
df.head(5)
```

```python
df.info()
```

```python
#remove all rows where Istrading is false
crypto_false = df[df["IsTrading"] == False].index
df.drop(crypto_false, inplace=True)
df.info()
```

```python
#remove IsTrading Column
df.drop('IsTrading', axis=1, inplace=True)
df.info()
```

```python
#drop rows with Nan values
df.dropna(inplace=True)
df.info()
```

```python
#filter TotalCoinsMined that are >0
df_filtered = df[df['TotalCoinsMined'] > 0]
df_filtered.info()
```

```python
df.drop('CoinName', axis=1, inplace=True)
df.info()
```

```python
df_filtered.drop('CoinName', axis=1, inplace=True)
df_filtered.info()
```

```python
df_filtered.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_filtered.drop_duplicates(inplace=True)
```

```python
print(f'Duplicate Entries: {df_filtered.duplicated().sum()}')
```
