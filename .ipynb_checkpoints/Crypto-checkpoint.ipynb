```python
import pandas as pd
from pathlib import Path
```

```python
#loading data
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path)
df.head(5)
```

```python
df.info()
```

```python
#remove all rows where Istrading is false
crypto_false = df[df["IsTrading"] == False].index
df.drop(crypto_false, inplace=True)
df.info()
```

```python
#remove IsTrading Column
df.drop('IsTrading', axis=1, inplace=True)
df.info()
```

```python
#drop rows with Nan values
df.dropna(inplace=True)
df.info()
```

```python
#filter TotalCoinsMined that are >0
df_filtered = df[df['TotalCoinsMined'] > 0]
df_filtered.info()
```

```python
df.drop('CoinName', axis=1, inplace=True)
df.info()
```

```python
df_filtered.drop('CoinName', axis=1, inplace=True)
df_filtered.info()
```

```python
df_filtered.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_filtered.drop_duplicates(inplace=True)
```

```python
print(f'Duplicate Entries: {df_filtered.duplicated().sum()}')
```

```python
df_filtered
```

```python
import numpy as np
crypto_clean = pd.get_dummies(df_filtered, columns=['Algorithm', 'ProofType'])
```

```python
crypto_clean
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(crypto_clean)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.9)

crypto_pca = pca.fit_transform(scaled_df)
crypto_pca
```

```python
crypto_pca_df = pd.DataFrame(data = crypto_pca)
crypto_pca_df.head()
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(learning_rate=35)
```

```python
tsne_features = tsne.fit_transform(crypto_pca_df)
```

```python
tsne_features.shape
```

```python
crypto_pca_df['x'] = tsne_features[:,0]

crypto_pca_df['y'] = tsne_features[:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(crypto_pca_df['x'], crypto_pca_df['y'])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Based on the analysis above, the Crypto currencies could be clustered into 4 groups.
